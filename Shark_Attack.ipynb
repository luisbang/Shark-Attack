 > **1.1.1) Conforme a resposta anterior qual é a atividade que teve mais casos fatais?**

```python
df_fatal_australia=df_fatal.loc[df_fatal['Country']=='AUSTRALIA',:]
df_fatal_australia=df_fatal_australia.reset_index()
df_fatal_australia
```

```python
df_activity=df_fatal_australia['Activity'].fillna('0')
df_activity
```

```python
t=[]
for i in list(df_activity):
    if re.search('[Bb][Aa][Tt][Hh][Ii][Nn][Gg]',i):
        t.append('Bathing')
    elif re.search('[Ss][Ww][Ii][Mm][Mm][Ii][Nn][Gg]',i):
        t.append('Swimming')
    elif re.search('[Dd][Ii][Vv][Ii][Nn][Gg]',i):
        t.append('Diving')
    elif re.search('[Ff][Ii][Ss][Hh][Ii][Nn][Gg]',i):
        t.append('Fishing')
    elif re.search('[Ss][Uu][Rr][Ff][Ii][Nn][Gg]',i):
        t.append('Surfing')
    elif re.search('[Ff][Ee][Ll][Ll]',i):
        t.append('Fell')
    elif re.search('[Ss][Tt][Aa][Nn][Dd][Ii][Nn][Gg]',i):
        t.append('Standing')
    elif re.search('[Ff][Ll][Oo][Aa][Tt][Ii][Nn][Gg]',i):
        t.append('Floating')
    elif re.search('[Ss][Hh][Ii][Pp][Ww][Rr][Ee][Cc][Kk]',i):
        t.append('Shipwreck')
    elif re.search('[Tt][Rr][Ee][Aa][Dd][Ii][Nn][Gg]',i):
        t.append('Treading')
    elif re.search('[Ww][Aa][Dd][Ii][Nn][Gg]',i):
        t.append('Wading')
    else:
        t.append(i)
t
```

```python
activity=pd.DataFrame(t).rename({0:'Activity_tratado'},axis=1)
activity
```

```python
df_fatal_australia_activity=pd.concat([df_fatal_australia,activity],axis=1).drop('index',axis=1)
df_fatal_australia_activity
```

```python
df_fatal_australia_activity['Activity_tratado'].value_counts()
```

```python
total_case=len(df_fatal_australia_activity)
total_case
```

```python
case_swimming=len(df_fatal_australia_activity.loc[df_fatal_australia_activity['Activity_tratado']=='Swimming',:])
case_swimming
```

```python
swimming_percentage=case_swimming/total_case*100
swimming_percentage
```
